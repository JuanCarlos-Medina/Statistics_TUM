# Working with DataFrames

```r
pokemon.df <- read.csv("pokemon.csv")
pokemon.df
```

```r
nrow(pokemon.df)
ncol(pokemon.df)
```

```r
colnames(pokemon.df)
```

Information on the type of data per column

```r
str(pokemon.df)
```

R thinks that Generation is an integer (continuous variable) but it is actually a factor (categorical variable)

```r
pokemon.df$Generation = as.factor(pokemon.df$Generation)
```

R thinks that Name is a categorical value but it is actually simply text (characters)

```r
pokemon.df$Name = as.character(pokemon.df$Name)
```

```r
str(pokemon.df)
```

## Ploting

We will use ggplot2 library to create nice visualizations.

```r
library(ggplot2)
```

### Frequency Distribution

```r
ggplot(data=pokemon.df, aes(Attack)) + geom_histogram()
```

```r
ggplot(data=pokemon.df, aes(Attack)) + geom_histogram(bins=20, color="black", fill="purple") 
```

### Density Distribution

```r
ggplot(data=pokemon.df, aes(Attack))  + geom_histogram(bins=20, color="black", fill="purple", 
                                                           aes(y = ..density..)) + geom_density(color="red")
```

### Cumulative Distribution

```r
ggplot(pokemon.df, aes(Attack)) + stat_ecdf(colour = "green")
```

### Bar Plot

```r
ggplot(pokemon.df, aes(Type.1)) +  geom_bar()
```

### Scatterplot

```r
ggplot(data=pokemon.df, aes(Attack, Defense)) + geom_point()
```

```r
ggplot(data=pokemon.df, aes(Attack, Defense, colour = Type.1)) + geom_point()
```

### Box Plot

```r
ggplot(pokemon.df, aes(x = "", y =Attack)) + geom_boxplot(fill = "#FFDB6D", color = "#C4961A")
```

```r
ggplot(pokemon.df, aes(x = Type.1, y =Attack)) + geom_boxplot(fill = "#FFDB6D", color = "#C4961A")
```

## Activity 5: Calculate the Pokemon Score (PS) for all pokemons

\begin{equation*}
PS = \frac{(\sqrt{Attack} + Defense^2) }{(HP-5)}
\end{equation*}

```r
pokemon.df$PS <- (sqrt(pokemon.df$Attack) + (pokemon.df$Defense)^2)/ (pokemon.df$HP - 5)
pokemon.df 
```
