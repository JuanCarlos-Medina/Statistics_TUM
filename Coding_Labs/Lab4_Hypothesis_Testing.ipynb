# <font color='#FD7C6E'>Normal Distribution and Critical Values </font>

In lab 2 we learned the following functions for the normal curve:
* **dnorm:** height of the probability distribution
* **pnorm:** probability of the distribution before a point
* **rnorm:** create random numbers from the normal distribution (sample)

Now we learn another VERY important function **qnorm**

qnorm is the opposite of pnorm. You give probabilities (or area below the curve) and it gives you the value for which

![alt text](norms.png "Title")

```r
qnorm(0.025)
qnorm(0.975)
```

We have seen these values before! The 95% area of the normal curve is between these values. We use these values for the two-sided hypothesis.

```r
qnorm(0.005)
qnorm(0.995)
```

### Activity 1: What is the critical value for the normal distribution if we have only a one sided hypothesis and 95% confidence? 

```r
qnorm(0.95)
```

We have used the standard normal distribution, it is also possible to use the previous functions on any normal distribution

```r
qnorm(0.025, mean=100, sd=10)
qnorm(0.975, mean=100, sd=10)
```

# <font color='#FD7C6E'>t Distribution </font>

Surprise! The same functions for the t distribution:

* **dt**
* **pt**
* **rt**
* **qt**

Lets copy code from Lab 2 for ploting:

```r
library(ggplot2)
options(repr.plot.width=4, repr.plot.height=4) # Size of the plots

create_plot <- function(x, y) {
   mydataframe <- data.frame(x,y)
   ggplot(data=mydataframe, aes(x,y)) + geom_jitter(color="#FF8E72")
}
```

dt helps us plot different t Distributions with different degrees of freedom

```r
x <- seq(-4,4,by=0.3) # Creates a sequence from -5 to 5 in spaces of 0.3
y <- dt(x, df=5)
create_plot(x,y)
```

```r
x <- seq(-4,4,by=0.3) # Creates a sequence from -5 to 5 in spaces of 0.3
y <- dt(x, df=10)
create_plot(x,y)
```

```r
x <- seq(-4,4,by=0.3) # Creates a sequence from -5 to 5 in spaces of 0.3
y <- dt(x, df=50)
create_plot(x,y)
```

### Activity 2: What is the critical values for a t distribution with 18 degrees of freedom, 95% confidence and a two sided hypothesis?

```r
qt(0.025, df=18)
qt(0.975, df=18)
```

# <font color='#FD7C6E'>p values </font>

p values refer to the area below the curve after the **test statistic**. We use therefore pnorm or pt to calculate them

### Activity 3: What is the p-value of a t test with 20 degrees of freedom and a t statistic of 1.8?

```r
1 - pt(1.8, df=20)
```
